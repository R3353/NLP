```
adjudications = pd.read_csv('adjudications.csv')
```

```
#find avg word count of each label
adjudications["Number of Words"] = adjudications["text"].apply(lambda n: len(n.split()))
adjudications.drop(['adjudicated', 'text', 'data point ID'], axis = 1).groupby(['label']).mean()
```

```
#punctuations 
text = adjudications['text']
puncts = text.str.findall(r'[^\w\s]')
adjudications["Punctuations"] = puncts.apply(len)
adjudications[['label', 'Punctuations']].groupby('label').mean()
```
